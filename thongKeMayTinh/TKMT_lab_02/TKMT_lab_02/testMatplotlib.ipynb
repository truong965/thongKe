```python

import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
arrX = np.array([-3, -2, 0, 5])
arrY = np.array([1, 3, 2, 10])
plt.plot(arrX,arrY)
plt.xlim(-5,7)
plt.ylim(0,15)
plt.show()
```

```python
x = np.linspace(-2, 2, 10)
y = x**2
z= x**3
#plt.subplots(figsize=(12, 4))
#plt.title('Hàm bình phương')
#plt.xlabel('x')
#plt.ylabel('y=x^2')
#r-- = red dash line, b--= blue dash line

plt.plot(x, y, 'r--')
plt.plot(x, x, 'b:')
plt.plot(x, z, 'g^')

plt.grid(True)
plt.show()
```

```python
x = np.linspace(-1.5, 1.5, 30)
plt.subplot(2, 2, 1)
plt.plot(x, x)         # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ nhất trái trên
  
plt.subplot(2, 2, 2)
plt.plot(x, x**2)      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ hai phải trên

plt.subplot(2, 2, 3)   # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ ba trái dưới
plt.plot(x, x**3)

plt.subplot(2, 2, 4)
plt.plot(x, x**4)      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ tư phải dưới

plt.show()
```

```python
x = np.linspace(-1.5, 1.5, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.text(0, 0, "Hàm số x", fontsize=15, color='blue', horizontalalignment="center",rotation=40)

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
pX=0.5
pY= pX**2
plt.plot(x, x**2)
plt.plot(pX,pX,'ro')
plt.text(0, 1, "Hàm số \n$y = x^2$", color='blue', horizontalalignment="center")
plt.text(pX, pY, "x = %0.2f\ny = %0.2f"%(pX, pY), color='gray')


plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show();
```

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.5
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Hàm số \n$y = x^2$", fontsize=15, color='blue', horizontalalignment="center")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), color='gray')

plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 50)

plt.plot(x, x**2, "r--", label="Square function")
plt.plot(x, x**3, "b-", label="Cube function")

plt.legend(loc="lower right")
plt.show()
```

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(2,1,1)
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(2,1,2)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.8)

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 3
data3 = np.random.randn(100) + 6
data4 = np.random.randn(100) + 10
plt.hist(data1, bins=5, color='g', alpha=0.75, label='bar hist') # default histtype='bar'
plt.hist(data2, color='b', alpha=0.1, histtype='stepfilled', label='stepfilled hist')
plt.hist(data3, color='r',histtype='barstacked', label='barstacked hist')
plt.hist(data4, color='purple',histtype='step', label='step hist')

plt.legend()
plt.show()
```

```python
for color in ['red', 'green', 'blue']:
    n = 10
    x, y = np.random.rand(2, n)
    scale = np.random.rand(n)
    scale = 500 * scale ** 2
    plt.scatter(x, y,s=scale,c=color,alpha=0.3)
plt.show()
```
