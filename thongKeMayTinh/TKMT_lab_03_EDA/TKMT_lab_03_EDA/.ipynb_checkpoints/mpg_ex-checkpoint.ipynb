# Phân tích mức tiêu hao nhiên liệu

## 1. ĐỌC DỮ LIỆU

Dữ liệu `mpg.csv` gồm các thông tin như sau:

- **mpg**: Miles/(US) gallon số gallon xăng trên 1 dặm
- **cylinders:** số lượng xilanh
- **displacement:**  Displacement/ dung tích xilanh (cu.in / $inches^3$)
- **horsepower:** công suất theo mã lực
- **weight:** trọng lượng (pound)
- **acceleration:** Gia tốc
- **model_year:** Năm sản xuất
- **origin:** Xuất xứ
- **name:** Tên dòng xe


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

## 2. TÌM HIỂU DỮ LIỆU

### 2.1 Kiểm tra có dòng nào không có dữ liệu?

```python
# mpg.info()
# mpg.describe(include='all')
mpg.isna().sum()
mpg[mpg.isnull().any(axis=1)]
```

### 2.2 Xóa các dòng không có dữ liệu


```python
mpg = mpg.dropna()
mpg.isna().sum()
```

### 2.3 Tìm trung binh, trung vị của các thuộc tính: 'cylinders', 'horsepower', 'weight'

```python
print('Trung bình')
print(mpg[['cylinders','horsepower','weight']].mean())
print('\nTrung vị')
mpg[['cylinders','horsepower','weight']].median()
```

### 2.4 Tìm miền giá trị và miền phân vị của thuộc tính 'mpg'

```python
print('Miền giá trị: ( ', mpg['mpg'].min() ,',', mpg['mpg'].max() ,')')
print('Miền phân vị:\n', mpg['mpg'].quantile([0.25,0.5,0.75]))
print('Độ lệch chuẩn:',mpg['mpg'].std())
```

## 3. SO SÁNH MỨC TIÊU HAO NHIÊN LIỆU TRUNG BÌNH CỦA CÁC DÒNG XE DỰA TRÊN XUẤT XỨ

So sánh mức tiêu hao nhiên liệu trung bình của các dòng xe Mỹ, Nhật và Châu Âu

```python
mpg[['mpg','origin']].groupby(['origin']).mean()
```

## 4. VẼ HISTOGRAM TRỌNG LƯỢNG CÁC XE

```python
plt.hist(mpg['weight'],color='r',bins=10,label ='hist weight')
```

## 5. TÍNH TỶ LỆ CÁC DÒNG XE DỰA TRÊN XUẤT XỨ

```python
name_counts= mpg[['origin','name']].groupby(['origin','name']).size().reset_index(name='count')
total_counts= mpg[['origin','name']].groupby(['origin']).size().reset_index(name='total count')
merge = pd.merge(name_counts,total_counts, on ='origin')
merge['percentage'] = round((merge['count']/merge['total count'])*100,2)
print(merge)
```

```python
# mpg['model_year'] = pd.to_numeric(mpg['model_year'], errors='coerce')  
data70=mpg.loc[(mpg['model_year']>=70) & (mpg['model_year']<80),'horsepower']
data80=mpg.loc[(mpg['model_year']>=80) & (mpg['model_year']<90),'horsepower']
data= [data70,data80]
plt.boxplot(data)
plt.xticks([1, 2], ['Thập niên 70', 'Thập niên 80'])
# plt.ylabel('Số công suất')
plt.title("công suất thập niên 70 và 80")
plt.show()
```

## 6. VẼ BOXPLOT SO SÁNH CÔNG SUẤT CỦA CÁC DÒNG XE THẬP NIÊN 70, 80

Lưu ý: Các dòng xe thập niên 70 (1970 - 1979), thập niên 80 (1980 - 1989)

## 7. KHẢO SÁT MỨC TIÊU HAO NHIÊN LIỆU

### 7.1 Với số lượng xi lanh

Thử vẽ bằng đồ thị scatter plot.

```python

plt.scatter(mpg['cylinders'],mpg['mpg'])
plt.title("mối quan hệ giữa số lượng xi lanh và mức tiêu hao nhiên liệu")
plt.xlabel("số lượng xi lanh")
# plt.ylabel('Mức tiêu hao nhiên liệu (mpg)')
plt.show()
```

### 7.2 Với công suất theo mã lực

```python
plt.scatter(mpg['horsepower'],mpg['mpg'])
plt.title("mối quan hệ giữa công suất theo mã lực và mức tiêu hao nhiên liệu")
plt.xlabel("công suất theo mã lực")
plt.show()
```

### 7.3 Với gia tốc

```python
plt.scatter(mpg['acceleration'],mpg['mpg'])
plt.title("mối quan hệ giữa gia tốc và mức tiêu hao nhiên liệu")
plt.xlabel("gia tốc")
plt.show()
```

### 7.4 Với năm sản xuất

```python
plt.scatter(mpg['model_year'],mpg['mpg'])
plt.title("mối quan hệ giữa năm sản xuất và mức tiêu hao nhiên liệu")
plt.xlabel("năm sản xuất")
plt.xticks(mpg['model_year'])
plt.show()
```
